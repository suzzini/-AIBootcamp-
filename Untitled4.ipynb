```python
def preprocess (df):

  df_temp = df.copy()

  #state 처리
  df_temp.dropna(subset=["State"], how="any", axis=0, inplace=True) #state가 결측치인 행 제거
  df_temp = df_temp[df_temp["State"]=="IL"] #IL이 아닌 행이 2개 밖에 없으므로 IL이 아닌 행은 제거

  #검사 요일을 구하는 새로운 컬럼 생성
  df_temp["Inspection Date"] = df_temp['Inspection Date'].apply(pd.to_datetime)
  #df_temp['Inspection Date']=pd.to_datetime(df_temp['Inspection Date'], format='%Y-%m-%d %H:%M:%S', errors='raise') #위의 방식으로 하는 것이 더 간단 ! 
  df_temp['day_name']=df_temp['Inspection Date'].dt.day_name()
  df_temp["year"] = df_temp["Inspection Date"].dt.year
  df_temp["month"] = df_temp["Inspection Date"].dt.month

  #City 컬럼에 CHICAGO를 다르게 표기한 것들을 통일시키고, 시카고가 맞는지 아닌지를 나타내는 feature 추가
  df_temp['City']=df_temp['City'].replace('Chicago','CHICAGO').replace('CHICAGOCHICAGO','CHICAGO').replace('CHicago','CHICAGO').replace('chicago','CHICAGO').replace('CCHICAGO','CHICAGO').replace('312CHICAGO','CHICAGO').replace('CHICAGO.','CHICAGO').replace('CHICAGOI','CHICAGO')
  df_temp["City_Chicago"] = [1 if i=='CHICAGO' else 0 for i in df_temp['City']]

  #Violations이 결측치인 값은 경고를 안받았다고 해석할 수 있으므로 0으로 채움
  df_temp.loc[:, "Violations"] = df_temp.loc[:, "Violations"].fillna('0')

  #Rish컬럼을 보기 좋게 숫자형으로 바꿈
  df_temp.dropna(subset=["Risk"], how="any", axis=0, inplace=True)
  df_temp.drop(df_temp.loc[df_temp['Risk']=='All'].index, inplace=True)
  df_temp.replace('Risk 1 (High)', 1, inplace=True)
  df_temp.replace('Risk 2 (Medium)', 2, inplace=True)
  df_temp.replace('Risk 3 (Low)', 3, inplace=True)
  #df_temp['Risk']=df_temp['Risk'].replace('Risk 1 (High)',1).replace('Risk 2 (Medium)',2).replace('Risk 3 (Low)',3).replace('All',1).fillna(0)

  #혹시 모를 cardinality 줄이기
  df_temp["DBA Name"] = df_temp["DBA Name"].str.upper()

  # 필요한 features 선택
  features = ["DBA Name", "year", "month","day_name", "License #", "Facility Type", "City_Chicago", "Latitude", "Longitude",  
                "Risk", "Inspection Type", "Violations", "Inspection Fail"]
  df_feature = df_temp[features]

  return df_feature 
```

```python
train = preprocess(train)
test = preprocess(test)
```

### 데이터 셋 분할하기

```python
# hold out validation으로 진행
from sklearn.model_selection import train_test_split

target = 'Inspection Fail'
features = train.columns.drop([target]) #불필요한 컬럼 제외하여 데이터 셋 나누기

train,val = train_test_split(train, test_size = 0.2, random_state = 2, stratify = train[target])


X_train = train[features]
y_train = train[target]

X_val = val[features]
y_val = val[target]

X_test = test[features]
y_test = test[target]
```

### 모델 학습을 위한 파이프라인을 구축 

```python
from sklearn.pipeline import make_pipeline
from category_encoders import TargetEncoder, OrdinalEncoder
from xgboost import XGBClassifier

pipe = make_pipeline(
    OrdinalEncoder(),
#     SimpleImputer(strategy='median'), # xgboost 에서는 missing value 처리를 해 줍니다
    XGBClassifier(n_estimators=200 
                  , random_state=2
                  , n_jobs=-1
                  , max_depth=7
                  , learning_rate=0.2
                 )
)

pipe.fit(X_train, y_train);
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score, roc_auc_score, classification_report, roc_curve

y_pred = pipe.predict(X_val)
print('검증 정확도: ', accuracy_score(y_val, y_pred))

print(classification_report(y_pred, y_val))
```

```python
encoder = OrdinalEncoder()
X_train_encoded = encoder.fit_transform(X_train) # 학습데이터
X_val_encoded = encoder.transform(X_val) # 검증데이터

model = XGBClassifier(
    n_estimators=1000,  # <= 1000 트리로 설정했지만, early stopping 에 따라 조절됩니다.
    max_depth=7,        # default=3, high cardinality 특성을 위해 기본보다 높여 보았습니다.
    learning_rate=0.2,
    #scale_pos_weight=ratio, # imbalance 데이터 일 경우 비율을 적용합니다.
    n_jobs=-1
)

#검증데이터가 필요한 이유 : validation score가 올라가는게 똑같아 그럼 멈춤!! 그래서 validationdata가 필요
#반복 횟수 만큼 변화가 없다 그럼 멈춤 . 검증데이터가 있어야 확인할 수 있음 

eval_set = [(X_train_encoded, y_train), 
            (X_val_encoded, y_val)]


#변화 없는 상태가 50번 반복된다 O, 변화가 없으면 멈추겟다

model.fit(X_train_encoded, y_train, 
          eval_set=eval_set,
          eval_metric='error', # #(wrong cases)/#(all cases)
          early_stopping_rounds=50
         ) # 50 rounds 동안 스코어의 개선이 없으면 멈춤
```

```python
print('검증 정확도', model.score(X_val_encoded, y_val))

print(classification_report(y_val, model.predict(X_val_encoded)))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import make_pipeline
from category_encoders import TargetEncoder, OrdinalEncoder
from xgboost import XGBClassifier

# XGBoost 분류기 생성
pipe = make_pipeline(
    OrdinalEncoder(),
#     SimpleImputer(strategy='median'), # xgboost 에서는 missing value 처리를 해 줍니다
    XGBClassifier(
                   random_state=2
                  , n_jobs=-1 
                  #, scale_pos_weight=ratio
                 )
)

# 초모수 격자생성
# 튜닝할 하이퍼파라미터의 범위를 지정해 주는 부분
dists = {
          'xgbclassifier__max_depth' : range(3,7,1),
           'xgbclassifier__learning_rate' : [0.01,0.2,0.05],
          'xgbclassifier__n_estimators' : [100,1000]
}

clf = RandomizedSearchCV(
    pipe, 
    param_distributions=dists, 
    n_iter=50, #50번 반복, 3회는 전체 150번의 학습/ 이 두 수치에 유의해야함
    cv=3,
    scoring='roc_auc',
    verbose=1,
    n_jobs=-1
)

clf.fit(X_train, y_train);
```

```python
print('최적 하이퍼파라미터: ', clf.best_params_)
print('roc_auc: ', clf.best_score_)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

y_pred = clf.predict_proba(X_test)[:, -1]
print('AUC score: ', roc_auc_score(y_test, y_pred))
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.scatter(fpr, tpr, color='blue')
plt.plot(fpr, tpr, color='green')
plt.title('ROC curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
```
